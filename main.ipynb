# Configuración Inicial de Librerias

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid", palette="Set2")
plt.rcParams["figure.figsize"] = (7,4)
print("Librerias cargadas correctamente")
```

# Carga del Archivo y conversión a Dataframe

```python
nombre_archivo = "Mini_Proyecto_Clientes_Promociones.xlsx"
df = pd.read_excel(nombre_archivo)
print("El archivo se procesó correctamente")
print(f"Filas: {df.shape[0]} / Columas: {df.shape[1]}")
df.head()
```

# Información general del Dataset

```python
print(" Información general del conjunto de datos:")
display(df.info())

print("\n Resumen estadístico de las columnas numéricas:")
display(df.describe())

print("\n Distribución de valores únicos por columna:")
for col in df.columns:
    print(f"{col}: {df[col].nunique()} valores únicos")
```

# Distribución por Genero

```python
print(" Analizamos la cantidad de clientes por género:")

sns.countplot(data=df, x="Genero", palette="pastel")
plt.title("Distribución por género")
plt.xlabel("Género")
plt.ylabel("Cantidad de clientes")
plt.show()

print("""
Observación:
Este gráfico nos permite ver si existe un equilibrio entre clientes
de distintos géneros o si hay predominio de alguno.
""")
```

# Distribución de Edades

```python
print("📊 Distribución de edades:")

sns.histplot(data=df, x="Edad", kde=True, bins=15)
plt.title("Distribución de Edades")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()

print("""
🧩 Observación:
Podemos identificar el rango etario predominante en la base de datos,
y si la distribución es más joven, adulta o madura.
""")
```


# Recompra por género

```python
print("Relación entre recompra y género:")

sns.countplot(data=df, x="Recompra", hue="Genero")
plt.title("Recompra por género")
plt.xlabel("¿Recompra?")
plt.ylabel("Cantidad de clientes")
plt.show()

print("""
Observación:
Permite analizar si existe una diferencia en el comportamiento de recompra
según el género de los clientes.
""")
```

# Relación entre ingreso mensual y total de compras



```python
print("Relación entre ingreso mensual y total de compras:")

sns.scatterplot(data=df, x="Ingreso_Mensual", y="Total_Compras", hue="Genero", alpha=0.8)
plt.title("Ingreso mensual vs Total de compras")
plt.xlabel("Ingreso mensual")
plt.ylabel("Total de compras")
plt.show()

print("""
Observación:
Nos ayuda a entender si el poder adquisitivo (ingreso mensual)
influye directamente en el volumen total de compras realizadas.
""")
```

# Promedio del monto promocional según recompra

```python
print("Promedio del monto promocional según recompra:")

sns.barplot(data=df, x="Recompra", y="Monto_Promo", palette="muted", ci=None)
plt.title("Monto promedio de promoción según recompra")
plt.xlabel("¿Recompra?")
plt.ylabel("Monto promedio")
plt.show()

print("""
Observación:
Si los clientes que recompran recibieron mayores montos promocionales,
podría indicar que las promociones fomentan la recompra.
""")
```

# Clientes que reciben promoción y recompra

```python
print("Clientes que reciben promoción y recompra:")

sns.countplot(data=df, x="Recibio_Promo", hue="Recompra", palette="coolwarm")
plt.title("Clientes que reciben promoción y recompra")
plt.xlabel("¿Recibe promoción?")
plt.ylabel("Cantidad de clientes")
plt.show()

print("""
Observación:
Permite analizar si las promociones están generando recompra
entre los clientes que las reciben.
""")
```

# Correlación entre variables numéricas

```python
print("Matriz de correlación entre variables numéricas:")

corr = df.select_dtypes(include='number').corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
plt.title("Matriz de correlación")
plt.show()

print("""
Observación:
Una correlación alta (positiva o negativa) indica una relación fuerte
entre dos variables numéricas, como Ingreso_Mensual y Total_Compras.
""")

print("\n Análisis exploratorio finalizado correctamente.")
```
